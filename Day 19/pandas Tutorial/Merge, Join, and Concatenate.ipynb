Merge, Join, and Concatenate

```python
import pandas as pd
df1=pd.read_csv(r"C:\Users\97798\Data\Day 19\pandas Tutorial\LOTR.csv")
df2=pd.read_csv(r"C:\Users\97798\Data\Day 19\pandas Tutorial\LOTR 2.csv")
df1
```

```python
df2
```

```python
#Merge
#It basically work as inner join
df1.merge(df2)
```

```python
df1.merge(df2,how='inner',on=['FellowshipID','FirstName'])
```

```python
df1.merge(df2,how='left',on=['FellowshipID','FirstName'])
```

```python
df1.merge(df2,how='right')
```

```python
#Provide us all the values regardless of the common
df1.merge(df2,how='outer')
```

```python
#cross merge(every possible value)
df1.merge(df2,how='cross')
```

```python
#joins
df1.join(df2, on='FellowshipID',how='outer',lsuffix='_left',rsuffix='_right')
```

```python
#best way to work with join by setting the index
df3=df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix='_left',rsuffix='_right',how='outer')
df3
```

```python
#Concatenate
# putting one dataframe on top of other rather than next to each other
```

```python
# only joining the common column
pd.concat([df1,df2],join='inner',axis=1)
```

```python
#append
#Not available in todays version
df1.append(df2)
```
