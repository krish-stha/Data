List of largest private companies

```python
import requests
from bs4 import BeautifulSoup

url="https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.text,'html')
soup.find_all('table')[1]
```

```python
table=soup.find_all('table')[1]
table_title=table.find_all('th')
table_title_final=[title.text.strip() for title in table_title]
print(table_title_final)
```

```python
import pandas as pd
df=pd.DataFrame(columns=table_title_final)
```

```python


rows=table.find_all('tr')
for row in rows[1:]:
    row_datas=row.find_all('td')
    individual_row_data=[row_data.text.strip()for row_data in row_datas]
   
    length=len(df)
    df.loc[length]=individual_row_data
    



```
