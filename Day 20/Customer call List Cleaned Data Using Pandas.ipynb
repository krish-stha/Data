Data Cleaning in Pandas

```python
import pandas as pd
df=pd.read_excel(r"C:\Users\97798\Data\Day 20\Customer Call List.xlsx")
df
```

```python
#removing Duplicates
df=df.drop_duplicates()
df
```

```python
df.drop(columns='Not_Useful_Column',inplace=True)
```

```python
df
```

```python
#removing unnecessary symbol of last_name
# df['Last_Name']=df['Last_Name'].str.lstrip('...')
# df['Last_Name']=df['Last_Name'].str.rstrip('_')
# df['Last_Name']=df['Last_Name'].str.lstrip('/')
# Best way to Do
df['Last_Name']=df['Last_Name'].str.strip('._/')
df
```

```python
df
```

```python
#Cleaning Phone Number
# df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
# df['Phone_Number']=df['Phone_Number'].apply(lambda x:str(x))#converting to string for indexing
# df['Phone_Number']=df['Phone_Number'].apply(lambda x:x[0:3]+'-'+x[3:6]+'-'+x[6:10])

# df['Phone_Number']=df['Phone_Number'].str.replace('Na--','')
# df['Phone_Number']=df['Phone_Number'].str.replace('nan--','')
df['Phone_Number']=df['Phone_Number'].str.replace('--','')
df

```

```python
# To Add new column
df[['Street Address','State','Zip_code']]=df['Address'].str.split(',',n=2,expand=True)
df
```

```python
df
```

```python
# For  paying customer and Do_Not_Contact
df[['Paying Customer','Do_Not_Contact']]=df[['Paying Customer','Do_Not_Contact']].replace('Yes','Y')
df[['Paying Customer','Do_Not_Contact']]=df[['Paying Customer','Do_Not_Contact']].replace('No','N')

df
```

```python
df=df.replace('N/a','')

# df.replace('NaN','') # cannot do this
# all the none and nan is replaced to blank by using fillna
df=df.fillna('')
df

```

```python
df
```

```python
# drop down who said do not contact
# looping through index
for i in df.index:
    #going through the value
    if df.loc[i,"Do_Not_Contact"] == 'Y':
        df.drop(i,inplace=True)
df
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].replace('','N')
df
```

```python
# removing phone number because without it , we cannot contact them

for x in df.index:
    if df.loc[x,'Phone_Number']=='':
        df.drop(x,inplace=True)
#Easy way to do
# df.dropna(subset='Phone_Number',inplace=True)
```

```python
df
```

```python
# df.reset_index(inplace=True)
df=df.reset_index(drop=True)
# df=df.drop(columns='index')
```

```python
df=df.drop(columns='Zip_code')
```
