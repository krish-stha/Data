```python
import pandas as pd

df=pd.read_excel(r"C:\Users\97798\Data\Day 20\Raw Data Customer Call List.xlsx")
df
```

```python
# Removing Duplicates
df=df.drop_duplicates()
df
```

```python
# Remove Unnecessary Column
df.drop(columns='Not_Useful_Column',inplace=True)
```

```python
#Cleaning last name
df['Last_Name']=df['Last_Name'].str.strip('_/.')
```

```python
#cleaning Phone number
df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True)
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x:str(x))
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x:x[0:3]+"-"+x[3:6]+"-"+x[6:10])
```

```python
df['Phone_Number']=df['Phone_Number'].replace('nan--','')
```

```python
df['Phone_Number']=df['Phone_Number'].replace('Na--','')
```

```python
pd.options.mode.copy_on_write = True 
```

```python
# Woking on Address
df[['Street','State','Zip_Code']]=df['Address'].str.split(',',n=2,expand=True)
```

```python
# Replacing yes with y and No with N in both column
df[['Paying Customer','Do_Not_Contact']]=df[['Paying Customer','Do_Not_Contact']].replace('Yes','Y')
```

```python
df[['Paying Customer','Do_Not_Contact']]=df[['Paying Customer','Do_Not_Contact']].replace('No','N')
```

```python
df=df.fillna('')
```

```python
df=df[df['Phone_Number'] != '']
```

```python
df=df[df['Do_Not_Contact'] == 'N']
```

```python
df=df.drop(columns='Zip_Code')
```

```python
df=df.drop_duplicates()
```
