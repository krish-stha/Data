```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\97798\Data\Day 20\world_population.csv")
```

```python
#to set decimal point to 2
pd.set_option('display.float_format',lambda x:'%.2f' %x) 
```

```python
df.describe()
```

```python
# To know the sum of missing value
df.isnull().sum()
```

```python
#  To know unique value
df.nunique()
```

```python
#Top 10 by 2022 population
df.sort_values(by="2022 Population",ascending=False).head(10)
```

```python
# Correaltion to each other
df.corr(numeric_only=True)
```

```python
#annot will show data label
sns.heatmap(df.corr(numeric_only=True),annot=True)
plt.rcParams['figure.figsize']=(10,5)
plt.show()
```

```python
df.groupby('Continent')[['2022 Population',
       '2020 Population', '2015 Population', '2010 Population',
       '2000 Population', '1990 Population', '1980 Population',
       '1970 Population']].mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)
```

```python
df.columns
```

```python
df2=df.groupby('Continent')[df.columns[5:13]].mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)
df2_reversed = df2[df2.columns[::-1]] 
df2_reversed.plot()

```

```python
df3=df2_reversed.transpose()
plt.rcParams['figure.figsize']=(18,5)
df3.plot()
plt.show()
```
