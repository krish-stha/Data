Pandas Visulatization

```python
import pandas as pd
```

```python

import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\97798\Data\Day 20\Ice Cream Ratings.csv")
# setting index to Data
df.set_index('Date')
```

```python
print(plt.style.available)
plt.style.use('_classic_test_patch')
```

```python
#Default, it will provide line graph
df.plot(x='Date',kind='line',title='Ice Cream Ratings',xlabel='Daily Ratings',ylabel='Scores')
```

```python
df.plot(x='Date',y=['Flavor Rating','Texture Rating'],kind='barh',stacked=True)
```

```python
#scatter plot
#s represents size of the dots
#c Represents colour
df.plot(kind='scatter',x='Texture Rating',y='Overall Rating',s=100,c='Red')
```

```python
#histogram
df.plot(kind='hist')
```

```python
#Box Plot
df.boxplot()
```

```python
#Area Plot
df.set_index('Date').plot.area()
```

```python
#pie chart
df.set_index('Date')['Texture Rating'].plot(kind='pie',legend=True).legend(loc='upper right') 
```
